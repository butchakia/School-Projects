## Part 3: Working with Categorical Variables
We will be creating lists to specify the order for each of the three categorical variables.

```python
clarity_levels = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']
cut_levels = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
color_levels = ['J', 'I', 'H', 'G', 'F', 'E', 'D']
```

We will now use these lists to communicate to Pandas the correct order for the
levels of the three categorical variables. 

```python
diamonds['cut'] = pd.Categorical(diamonds['cut'], categories=['Fair', 'Good', 'Very Good', 'Premium', 'Ideal'], ordered=True)
diamonds['color'] = pd.Categorical(diamonds['color'], categories=['J', 'I', 'H', 'G', 'F', 'E', 'D'], ordered=True)
diamonds['clarity'] = pd.Categorical(diamonds['clarity'], categories=['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF'], ordered=True)
```

We will now create lists of named colors to serve as palettes to be used for
visualizations later in the notebook. 

```python
clarity_pal = ['red', 'orange', 'yellow', 'green', 'blue', 'indigo', 'purple', 'pink']
color_pal = ['brown', 'aqua', 'yellow', 'gold', 'blue', 'indigo', 'silver']
cut_pal = ['maroon', 'orange', 'gold', 'green', 'blue']
```

## Part 5: Scatterplots of Price Against Carat
We will explore the relationship between the price and carat attributes by generating scatter plots.

```python
plt.figure(figsize=[8, 6])
plt.scatter(diamonds['carat'], diamonds['price'], s=20, alpha=0.4, color='blue', edgecolors='none')
plt.xlabel('Carat Size')
plt.ylabel('Price (in $)')
plt.title('Relationship between Diamond Price and Carat Size')
plt.show()
```

We will reproduce the plot from above, but will now color the points according to
their clarity.

```python
plt.figure(figsize=[8, 6])
for clarity, color in zip(clarity_levels, clarity_pal):
    plt.scatter(diamonds[diamonds['clarity'] == clarity]['carat'], diamonds[diamonds['clarity'] == clarity]['price'],
                s=20, alpha=0.4, color=color, edgecolors='none', label=clarity)
plt.xlabel('Carat Size')
plt.ylabel('Price (in $)')
plt.title('Relationship between Diamond Price and Carat Size')
plt.legend()
plt.show()
```

We will now separate the points for each of the eight clarity levels into its own
subplot.

```python
fig, axes = plt.subplots(2, 4, figsize=[12, 6])
for clarity, color, ax in zip(clarity_levels, clarity_pal, axes.flatten()):
    subset = diamonds[diamonds['clarity'] == clarity]
    ax.scatter(subset['carat'], subset['price'], s=20, alpha=0.4, color=color, edgecolors='none')
    ax.set_xlabel('Carat Size')
    ax.set_ylabel('Price (in $)')
    ax.set_title(clarity)
    ax.set_xlim([0, 4])
    ax.set_ylim([0, 20000])
plt.tight_layout()
plt.show() 
```

## Part 6: Applying Logarithmic Transformations
We will apply logarithmic transformations to the price and carat columns, and will explore the distribution of
these transformed variables.

```python
diamonds['ln_carat'] = np.log(diamonds['carat'])
diamonds['ln_price'] = np.log(diamonds['price'])
```

We will now use histograms to explore the distribution of the diamond prices, and
the log of the diamond prices. 

```python
fig, axs = plt.subplots(1, 2, figsize=[12, 4])
axs[0].hist(diamonds['price'], bins=20, edgecolor='black', color='violet')
axs[0].set_xlabel('Price')
axs[0].set_ylabel('Count')
axs[0].set_title('Histogram of Prices')
axs[1].hist(diamonds['ln_price'], bins=20, edgecolor='black', color='pink')
axs[1].set_xlabel('Natural Log of Price')
axs[1].set_ylabel('Count')
axs[1].set_title('Histogram of Log-Prices')
plt.tight_layout()
plt.show()
```

We will now use histograms to explore the distribution of the carat sizes, and the
log of the carat sizes. 

```python
fig, axs = plt.subplots(1, 2, figsize=[12, 4])
axs[0].hist(diamonds['carat'], bins=20, edgecolor='black', color='red')
axs[0].set_xlabel('Carat Size')
axs[0].set_ylabel('Count')
axs[0].set_title('Histogram of Carat Sizes')
axs[1].hist(diamonds['ln_carat'], bins=20, edgecolor='black', color='orange')
axs[1].set_xlabel('Natural Log of Carat Size')
axs[1].set_ylabel('Count')
axs[1].set_title('Histogram of Log-Carat Sizes')
plt.tight_layout()
plt.show()
```

## Part 7: Scatterplot of Transformed Variables
We will create a scatterplot of ln_price against ln_carat, with points colored according to their
clarity.

```python
plt.figure(figsize=[8, 6])
for clarity, color in zip(clarity_levels, clarity_pal):
    plt.scatter(diamonds[diamonds['clarity'] == clarity]['ln_carat'], diamonds[diamonds['clarity'] == clarity]['ln_price'],
                s=20, alpha=0.4, color=color, edgecolors='none', label=clarity)
plt.xlabel('Natural Log of Carat Size')
plt.ylabel('Natural Log of Price')
plt.title('Relationship between Log-Price and Log-Carat Size')
plt.legend()
plt.show()
```

## Part 8: Grouping by Cut
We will create a grouped DataFrame displaying the mean price and carat size for each cut level. 

```python
gb_cut = diamonds[['cut', 'price', 'carat']].groupby('cut').mean()
gb_cut
```

We will now use bar charts to graphically display the information from the
DataFrame above.

```python
plt.figure(figsize=[12, 4])
plt.subplot(1, 2, 1)
plt.bar(gb_cut.index, gb_cut['price'], edgecolor='black', color=cut_pal)
plt.xlabel('Cut')
plt.ylabel('Mean Price')
plt.title('Mean Price by Cut')
plt.subplot(1, 2, 2)
plt.bar(gb_cut.index, gb_cut['carat'], edgecolor='black', color=cut_pal)
plt.xlabel('Cut')
plt.ylabel('Mean Carat Size')
plt.title('Mean Carat Size by Cut')
plt.tight_layout()
plt.show() 
```
